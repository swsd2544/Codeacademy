```python
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('all_data.csv')
print(df)
```

```python
# Compare GDP of each country
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(data=df.sort_values(['GDP', 'Country']), x='Country', y='GDP')
plt.show()
```

```python
#Plot another graph of Country GDP.
#Only Zimbabwe, Chile, Mexico
```

```python
temp_df = df[df.Country.isin(['Zimbabwe', 'Chile', 'Mexico'])]
plt.figure(figsize=(10, 8))
sns.barplot(data=temp_df.sort_values(['GDP', 'Country']), x='Country', y='GDP')
plt.show()
```

```python
#Find correlation between GDP and Life Expectancy
#Using the pearsonr from scipy.stats
```

```python
r, p = pearsonr(df.GDP, df['Life expectancy at birth (years)'])
print(r)
```

```python
# We can see that r value not that close to 1, 
# so it isn't linear relationship between gdp and life expectancy at birth
# To find out more, we use the scatterplot to show relationship between these two variables
```

```python
plt.figure(figsize=(10 ,8))
sns.scatterplot(data=df, x='GDP', y='Life expectancy at birth (years)', hue='Country')
plt.show()
```

```python
#Zimbabwe is an outlier
plt.figure(figsize=(10 ,8))
ax1 = plt.subplot(2, 1, 1)
sns.scatterplot(data=df[df.Country != 'Zimbabwe'], x='GDP', y='Life expectancy at birth (years)', hue='Country')
ax2 = plt.subplot(2, 1, 2)
sns.scatterplot(data=df[df.Country == 'Zimbabwe'], x='GDP', y='Life expectancy at birth (years)', hue='Country')
plt.show()
```

```python
# We can see that, in some country, there is a linear relationship between country GDP and life expectancy
# Then we will plot another graph that will show increment of GDP of each country over time
```

```python
plt.figure(figsize=(10, 8))
sns.lineplot(data=df, x='Year', y='GDP', hue='Country')
plt.show()
```

```python
#Zimbabwe is an outlier
plt.figure(figsize=(10 ,8))
ax1 = plt.subplot(3, 1, 1)
sns.lineplot(data=df[~df.Country.isin(['Zimbabwe', 'Chile', 'Mexico'])], x='Year', y='GDP', hue='Country')
ax2 = plt.subplot(3, 1, 2)
sns.lineplot(data=df[df.Country.isin(['Chile', 'Mexico'])], x='Year', y='GDP', hue='Country')
ax3 = plt.subplot(3, 1, 3)
sns.lineplot(data=df[df.Country == 'Zimbabwe'], x='Year', y='GDP', hue='Country')
plt.show()
```
